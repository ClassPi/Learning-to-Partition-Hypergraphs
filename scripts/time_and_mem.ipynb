```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
species = ("HGNN+", "HyperGCN", "GraphSAGE")
penguin_means = {
    'Cora': (71.6, 48*60+35, 42.7),
    'Pubmed': (7*60+39, 3800, 9*60+13.8),
    'Cooking200': (7*60+35, 32*60+35, 24*60+27.5),  
}

# 指定颜色
colors = ['#5fc8ee', '#cff3a8', '#efa05f']

x = np.arange(len(species))  # 标签位置
width = 0.25  # 柱子宽度
multiplier = 0

# 控制字体大小的变量
fontsize = 15

fig, ax = plt.subplots(layout='constrained')

# 绘制柱状图
for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    color = colors[multiplier]  # 按顺序取颜色
    rects = ax.bar(x + offset, measurement, width, label=attribute, color=color)

    # 在每个柱子上方显示以分钟为单位的标签
    for rect in rects:
        height = rect.get_height()

        # 如果高度为3800，显示为200分钟
        if height == 3800:
            label = '200'
        else:
            label = f'{height/60:.1f}'  # 将秒数转换为分钟显示

        if height == 3800:
            ax.text(rect.get_x() + rect.get_width() / 2, height - 200
                    , label,
                    ha='center', va='bottom', fontsize=fontsize, color='black')
        else:
            # 在柱子上方显示标签
            ax.text(rect.get_x() + rect.get_width() / 2, height, label,
                    ha='center', va='bottom', fontsize=fontsize)

    multiplier += 1

# 设置x轴和y轴标签，使用变量 fontsize 控制字体大小
ax.set_ylabel('Time (Minutes)', fontsize=fontsize,labelpad=-8)
ax.set_title('Time Efficiency', fontsize=fontsize)

# 设置 x 轴刻度的字体大小
ax.set_xticks(x + width, species, fontsize=fontsize)

# 设置图例字体大小
ax.legend(loc='upper right', ncols=1, fontsize=fontsize)

# 设置非等距的 y 轴刻度
ax.set_ylim(0, 3800)
yticks = [600, 1200, 1800, 2400, 3000, 3700]  # y 轴刻度
ytick_labels = ['10', '20', '30', '40', '50', '60+']  # y 轴刻度标签
ax.set_yticks(yticks)
ax.set_yticklabels(ytick_labels, fontsize=fontsize)

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
species = ("HGNN+", "HyperGCN", "GraphSAGE")
penguin_means = {
    'Cora':(164, 164, 236), 
    'Pubmed':(1376, 1396, 2820),
    'Cooking200': (1586, 1590, 2136), 
}

# 指定颜色
colors = ['#5fc8ee', '#cff3a8', '#efa05f']

x = np.arange(len(species))  # 标签位置
width = 0.25  # 柱子宽度
multiplier = 0

# 控制字体大小的变量
fontsize = 15

fig, ax = plt.subplots(layout='constrained')

# 绘制柱状图
for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    color = colors[multiplier]  # 按顺序取颜色
    rects = ax.bar(x + offset, measurement, width, label=attribute, color=color)

    # 在每个柱子上方显示以分钟为单位的标签
    for rect in rects:
        height = rect.get_height()
        label = height
        ax.text(rect.get_x() + rect.get_width() / 2, height, label,
                ha='center', va='bottom', fontsize=13)

    multiplier += 1

# 设置x轴和y轴标签，使用变量 fontsize 控制字体大小
ax.set_ylabel('VRAM (MB)', fontsize=fontsize,labelpad=4)
ax.set_title('VRAM Usage', fontsize=fontsize)

# 设置 x 轴刻度的字体大小
ax.set_xticks(x + width, species, fontsize=fontsize)

# 设置图例字体大小
ax.legend(loc='upper left', ncols=1, fontsize=fontsize)

# 设置非等距的 y 轴刻度
ax.set_ylim(0, 3000)
# yticks = [600, 1200, 1800, 2400, 3000, 3700]  # y 轴刻度
# ytick_labels = ['10', '20', '30', '40', '50', '60+']  # y 轴刻度标签
ax.tick_params(axis='y', labelsize=fontsize)
# ax.set_yticklabels(ytick_labels, fontsize=fontsize)

# 显示图表
plt.show()
```
